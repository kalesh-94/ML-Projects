```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import os
```

```python
data_set = pd.read_csv("/content/diabetes.csv")
```

```python
data_set.head()
```

```python
data_set.describe()
```

```python
data_set.info()
```

```python
data_set.isnull().sum()
```

```python
features_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
```

```python
x = data_set[features_cols].values
y = data_set.Outcome
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=16)
```

```python
logreg = LogisticRegression(max_iter=200)
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
print(logreg.predict([[6,148,72,35,0,37,99,0]]))
```

```python
import pickle
import os
```

```python
current_directory = os.getcwd()
```

```python
model_path = os.path.join(current_directory,"model.pkl")
```

```python
pickle.dump(logreg,open(model_path,'wb'))
```

```python
with open(model_path,'rb') as f:
    model = pickle.load(f)
```

```python
# Make predictions
predictions = model.predict([[6,148,72,35,0,37,99,0]])

# Convert numerical output to text
output_labels = ["You have diabetes" if pred == 1 else "You do not." for pred in predictions]

# Display results
print(output_labels)
```

```python
# Make predictions
predictions = model.predict([[6,148,72,35,0,37,0,0]])

# Convert numerical output to text
output_labels = ["You have diabetes" if pred == 1 else "You do not." for pred in predictions]

# Display results
print(output_labels)
```

```python
from sklearn.metrics import accuracy_score


y_pred = model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print accuracy
print(f"Model Accuracy: {accuracy:.2f}")
```
